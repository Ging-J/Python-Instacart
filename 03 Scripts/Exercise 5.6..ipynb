```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# File path and name
file_path = "/Users/ging/Documents/Data Analytics Course/Data Ethics Achivement 5 /01 Data/Original Data/5.6_Time_Series_Task.xlsx"

# Loading the data from the "Sales" sheet
df = pd.read_excel(file_path, sheet_name='Sales')

# Plotting the time series as a Line chart
plt.figure(figsize=(10, 6))
plt.plot(df['Time'], df['Retail_Business'], marker='o', linestyle='-')
plt.title('Monthly Revenue Time Series')
plt.xlabel('Time')
plt.ylabel('Retail Business ($)')
plt.grid(True)
plt.show()
```

```python
import pandas as pd

# Load the Excel file
file_path = "/Users/ging/Documents/Data Analytics Course/Data Ethics Achivement 5 /01 Data/Original Data/5.6_Time_Series_Task.xlsx"
sheet_name = "Oil"
df = pd.read_excel(file_path, sheet_name=sheet_name)

# Convert "Year" column to datetime format
df['Year'] = pd.to_datetime(df['Year'])

# Calculate the 5-year moving average
df['5 Yr MA'] = df['Avg Oil Price'].rolling(window=5).mean()

# Save the updated DataFrame to the Prepared Data folder
prepared_file_path = "/Users/ging/Documents/Data Analytics Course/Data Ethics Achivement 5 /01 Data/Prepared Data/5.6_Time_Series_Task_updated.xlsx"
df.to_excel(prepared_file_path, index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file
file_path = "/Users/ging/Documents/Data Analytics Course/Data Ethics Achivement 5 /01 Data/Prepared Data/5.6_Time_Series_Task.xlsx"
sheet_name = "Oil"
df = pd.read_excel(file_path, sheet_name=sheet_name)

# Convert "Year" column to datetime format
df['Year'] = pd.to_datetime(df['Year'])

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['Avg Oil Price'], marker='o', label='Average Oil Price')
plt.plot(df['Year'], df['5 Yr MA'], linestyle='--', color='red', label='5 Year Moving Average')
plt.xlabel('Year')
plt.ylabel('Average Oil Price ($)')
plt.title('Average Oil Price and 5 Year Moving Average')
plt.legend()
plt.grid(True)
plt.show()
```
